```python
X = df.drop('price', axis=1)

y = df['price']
```

```python

print('Shape of X = ', X.shape)

print('Shape of y = ', y.shape)
```

```python
"""
axis 0 and axis 1
axis 0 represents rows and
axis 1 represents columns
"""
```

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=51)
```

```python
print('Shape of X_train = ', X_train.shape)

print('Shape of y_train = ', y_train.shape)

print('Shape of X_test = ', X_test.shape)

print('Shape of y_test = ', y_test.shape)
```

```python
"""

Feature scaling

The most common techniques of feature scaling are Normalization and Standardization.

Machine learning algorithm just sees number â€” if there is a vast difference in the range say few ranging in thousands
and few ranging in the tens, it makes the underlying assumption that higher ranging numbers have superiority of some sort. 

So these more significant number starts playing a more decisive role while training the model.

"""
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

sc.fit(X_train)

```

```python
X_train = sc.transform(X_train)

X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X_train, y_train)
```

```python
lr.coef_
```

```python
lr.predict(X_test)
```

```python
for i in lr.predict(X_test):
    print(i)
```

```python
y_test
```

```python
lr.score(X_test, y_test)
```

```python
# Ridge and Lasso Regression ML Model

from sklearn.linear_model import Ridge, Lasso

rd = Ridge()

rd.fit(X_train,y_train)

```

```python
rd.score(X_test, y_test)
```

```python
ls = Lasso()

ls.fit(X_train,y_train)

ls.score(X_test, y_test)
```

```python
rd2 = Ridge(alpha = 2)

rd2.fit(X_train,y_train)

rd2.score(X_test, y_test)
```

```python
ls3 = Lasso(alpha=3)

ls3.fit(X_train,y_train)

ls3.score(X_test, y_test)
```
